## Hypothesis Test 1

**Does the MPAA rating of a movie (G/PG/PG-13/R) affect how much revenue the movie generates?**

$H_0$: There is no difference between the MPAA rating of a movie and the amount of revenue the movie generates.

$H_A$: There is a huge difference between the MPAA rating of a movie and the amount of revenue the movie generates. 

**Best Test:**
- With multiple categorical groups
- And one numeric column

**Anova is the best test**

**Checking for:**
- Normality
- Equal Variance
- No Significant Outliers

```python
reve_certi = df[['revenue', 'certification']]
```

```python
reve_certi.dropna(inplace = True)
```

```python
reve_certi
```

```python
value_counts = reve_certi['certification'].value_counts()

# Select the values where the count is less than 3 (or 5 if you like)
to_remove = value_counts[value_counts <= 3].index

# Keep rows where the city column is not in to_remove
reve_certi = reve_certi[~reve_certi.certification.isin(to_remove)]
```

```python
reve_certi['certification'].value_counts()
```

```python
#visualizing the variables
plt.figure(figsize = (20,7))
sns.barplot(data = reve_certi, x= 'certification', y = 'revenue')
```

```python
## Create groups dictionary. 
certi_groups = {}
## Loop through all unique categories
for i in reve_certi['certification'].unique():
    ## Get series for group and rename
    data = reve_certi.loc[reve_certi['certification']==i,'revenue'].copy()
    
    # save into the dictionary
    certi_groups[i] = data
    
certi_groups.keys()
```

### **Testing Assumptions**

**Assumption of Normality**

```python
## Running a normal test on each group and confirming there are >20 in each group
norm_results = {}
for i, data in certi_groups.items():
    stat, p = stats.normaltest(data)
    ## save the p val, test statistic, and the size of the group
    norm_results[i] = {'n': len(data),
                             'p':p,
                             'test stat':stat,}
## convert to a dataframe
norm_results_df = pd.DataFrame(norm_results).T
norm_results_df
```

```python
### checking sig with pandas 
norm_results_df['significance'] = norm_results_df['p'] <0.05
norm_results_df
```

**We see that the groups are not normally distributed. However, we have groups large enough that we can proceed without this condition being met.**

**Assumptions of Equal Variance**

```python
rc_equal_variance = stats.levene(*certi_groups.values())

print(rc_equal_variance)
print(rc_equal_variance.pvalue < 0.05)
```

**It does not have equal variance.**
**We will continue with the One way ANOVA Test**

```python
# perform the correct hypothesis test
# note again the use of the * operator for convenience
result = stats.f_oneway( *certi_groups.values())
print(result)

print(result.pvalue < 0.05)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
## save the values as kg_lost and the labels to the Diet
values = reve_certi['revenue']
labels = reve_certi['certification']
```

```python
## perform tukey's multiple comparison test and display the summary
tukeys_results = pairwise_tukeyhsd(values,labels)
tukeys_results.summary()
```

```python
#visualizing the variables
plt.figure(figsize = (20,7))
sns.barplot(data = reve_certi, x= 'certification', y = 'revenue')
```

### Interpretting the results

**With the Tukey test we see that there are 10 groups that reject the null hypothesis and 5 that do not**

**Looking at the One Way ANOVA test, it rejects the null hypothesis straight out.**

**The most significant difference is between the movie with NC-17 & R rated movies. There p-value is 0.71**

### Visual

**We can see that PG-13 movies got the highest amount of revenue out of all the certification types.**

### Conclusion
**We reject the null hypothesis**
**Therefore there is a significant difference between the certification of a movie and the amount of revenue that it generates.**

## Hypothesis Test 2

### **Does runtime have any affect on the revenue of a film.**

**$H_0$: There is not difference between the runtime of a film and its revenue**

**$H_A$: There is a significant difference between the runtime of a film and its revenue.**

**Technique:**
- Since runtime is a revenue is a numeric variable, we will turn it into a binary.
- The difference will be movie with a run time of less than 90 minutes and those with more than 90 minutes. 

**Best Test:**
- With two categorical groups
- And one numeric column

**2 Sample t-test is the best test**

**Checking for:**
- Any significant outliers
- Equal variance
- Normality

 

```python
df.head()
```

```python
df['runtime'].mean()
```

```python
df['runtime'].max()
```

```python
df['runtime'].min()
```

```python
df['runtime'].value_counts()
```

```python
df['more_than_90_min'] = df['runtime'] > 90

df
```

```python
rr_df = df[['more_than_90_min', 'revenue']]

rr_df.head()
```

```python
rr_df['more_than_90_min'].value_counts()
```

```python
#separating the col of interests based on the group.

less_than_90 = rr_df.loc[rr_df['more_than_90_min'] == False, ['revenue','more_than_90_min']]


more_than_90 = rr_df.loc[rr_df['more_than_90_min'] == True, ['revenue','more_than_90_min']]


print(f'{len(less_than_90)} have less than 90 minute runtime')

print(f'{len(more_than_90)} has more than 90 minute runtime')
```

```python
sns.barplot(data=rr_df, x='more_than_90_min',y='revenue');
```

```python
#checking for outliers
less_than_90_rev = less_than_90['revenue']
more_than_90_rev = more_than_90['revenue']
```

```python

import scipy.stats as stats

#outliers for less_than_90_rev
l_zscore = stats.zscore(less_than_90_rev)
l_outlier = abs(l_zscore) > 3
print(f' Movies less than 90 minutes has: {np.sum(l_outlier)} outliers')
```

```python

#outliers for more_than_90_rev
m_zscore = stats.zscore(more_than_90_rev)
m_outlier = abs(m_zscore) > 3
print(f' Movies more than 90 minutes has: {np.sum(m_outlier)} outliers')
```

```python
#getting rid of outliers

#less_than_90_rev 
less_than_90 = less_than_90_rev[(np.abs(stats.zscore(less_than_90_rev)) < 3)] 

#more_than_90_rev
more_than_90 = more_than_90_rev[(np.abs(stats.zscore(more_than_90_rev)) < 3)]

print(less_than_90)

print(more_than_90)
```

```python
# normality for recevied discount
normality_less_than_90 = stats.normaltest(less_than_90)

# normality for no discount
normality_more_than_90 = stats.normaltest(more_than_90)

print('Normality for Less Than 90 runtime:')
print(normality_less_than_90)
print("\n")
print('Is pvalue below alpha for Received Discount?')
print(normality_less_than_90.pvalue < 0.05)

print("\n")

print('Normality for No Discount')
print(normality_more_than_90)
print("\n")
print('Is pvalue below alpha for No Discount?')
print(normality_more_than_90.pvalue < 0.05)
```

**For both groups we can see that p value is below alpha. This means that neither group is normally distributed. However, we have a group large enough that we can proceed without this condition being met.**

### **Checking for Equal Variance**

```python
#equal variance
final_results = stats.levene(more_than_90, less_than_90)

print('Is normal variance below alpha?')
final_results.pvalue < 0.05
```

**The set does not have equal variance.**

**We will continue with the t-test. We will make sure that equal_var = False.**

### Performing t-test

```python
#t-test results
t_test_results = stats.ttest_ind(more_than_90, less_than_90, equal_var = False)

print(t_test_results)
t_test_results.pvalue < 0.05
```

```python
plt.figure(figsize = (15,6))
sns.barplot(data=rr_df, x='more_than_90_min',y='revenue');
```

## Interpretting the results

### Visual

**We can see that Movies with a runtime higher than 90 minutes does make more money than those that are less than 90 minutes.**

### Conclusion
**We reject the null hypothesis**

**Therefore there is a significant difference between the revenue of movies whose runtime is less than 90 minutes and those with runtime of more than 90 minutes.**
