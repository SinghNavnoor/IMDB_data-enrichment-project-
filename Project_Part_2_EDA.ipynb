#### The goal is to bring two datasets together.
#### If we merge, using imdb_id, it will work but them we will have double the columns.
#### Going to add a new Year column to both datasets and then using pd.concat going to merge them together. 
#### This will help with accurate numbers. 
#### Then continue with the EDA 

```python
final_year_2000['year'] = 2000
```

```python
final_year_2001['year'] = 2001
```

```python
final_year_2001.tail(2)
```

```python
both_years = pd.concat([final_year_2000, final_year_2001], ignore_index = True)
```

```python
both_years.columns
```

```python
both_years.info()
```

### How many movies had at least some valid financial information 

```python
some_value = both_years.loc[ (both_years['revenue'] > 0.0) | (both_years['budget'] > 0.0)]
```

```python
some_value
```

```python
print(f"{len(some_value)} movie in the combined dataset has some type of financial information.")
```

### How many movies are there in each of the certification categories (G/PG/PG-13/R)

#### Going to group by movies that have some financial value.
#### So I will be using the some_value variable for grouping. 

```python
cert_cat = some_value[['certification', 'imdb_id']]

cert_cat
```

```python
cert_cat.groupby('certification').count()
```

### What is the average revenue per certification category?

#### Using the variable some_value

```python
#selecting the two columns
avg_by_certification = some_value[['certification','revenue']]

avg_by_certification.head()
```

```python
avg_by_certification.groupby('certification').mean().round(1)
```

### What is the average budget per certification category?

#### Using the same strategy that I did for average revenue.

```python
avg_budget = some_value[['certification','budget']]
```

```python
avg_budget.groupby('certification').mean().round(1)
```
