## Hypothesis Test 1

**Does the MPAA rating of a movie (G/PG/PG-13/R) affect how much revenue the movie generates?**

$H_0$: There is no difference between the MPAA rating of a movie and the amount of revenue the movie generates.

$H_A$: There is a huge difference between the MPAA rating of a movie and the amount of revenue the movie generates. 

**Best Test:**
- With multiple categorical groups
- And one numeric column

**Anova is the best test**

**Checking for:**
- Normality
- Equal Variance
- No Significant Outliers

```python
reve_certi = df[['revenue', 'certification']]
```

```python
reve_certi.dropna(inplace = True)
```

```python
reve_certi
```

```python
value_counts = reve_certi['certification'].value_counts()

# Select the values where the count is less than 3 (or 5 if you like)
to_remove = value_counts[value_counts <= 3].index

# Keep rows where the city column is not in to_remove
reve_certi = reve_certi[~reve_certi.certification.isin(to_remove)]
```

```python
reve_certi['certification'].value_counts()
```

```python
#visualizing the variables
plt.figure(figsize = (20,7))
sns.barplot(data = reve_certi, x= 'certification', y = 'revenue')
```

```python
## Create groups dictionary. There are only 3 groups, but this is even more helpful when there are many groups
certi_groups = {}
## Loop through all unique categories
for i in reve_certi['certification'].unique():
    ## Get series for group and rename
    data = reve_certi.loc[reve_certi['certification']==i,'revenue'].copy()
    
    # save into the dictionary
    certi_groups[i] = data
    
certi_groups.keys()
```

### **Testing Assumptions**

**Assumption of Normality**

```python
## Running a normal test on each group and confirming there are >20 in each group
norm_results = {}
for i, data in certi_groups.items():
    stat, p = stats.normaltest(data)
    ## save the p val, test statistic, and the size of the group
    norm_results[i] = {'n': len(data),
                             'p':p,
                             'test stat':stat,}
## convert to a dataframe
norm_results_df = pd.DataFrame(norm_results).T
norm_results_df
```

```python
### checking sig with pandas 
norm_results_df['significance'] = norm_results_df['p'] <0.05
norm_results_df
```

**We see that the groups are not normally distributed. However, we have groups large enough that we can proceed without this condition being met.**

**Assumptions of Equal Variance**

```python
rc_equal_variance = stats.levene(*certi_groups.values())

print(rc_equal_variance)
print(rc_equal_variance.pvalue < 0.05)
```

**It does not have equal variance.**
**We will continue with the One way ANOVA Test**

```python
# perform the correct hypothesis test
# note again the use of the * operator for convenience
result = stats.f_oneway( *certi_groups.values())
print(result)

print(result.pvalue < 0.05)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
## save the values as kg_lost and the labels to the Diet
values = reve_certi['revenue']
labels = reve_certi['certification']
```

```python
## perform tukey's multiple comparison test and display the summary
tukeys_results = pairwise_tukeyhsd(values,labels)
tukeys_results.summary()
```

```python
#visualizing the variables
plt.figure(figsize = (20,7))
sns.barplot(data = reve_certi, x= 'certification', y = 'revenue')
```

### Interpretting the results

**With the Tukey test we see that there are 10 groups that reject the null hypothesis and 5 that do not**

**Looking at the One Way ANOVA test, it rejects the null hypothesis straight out.**

**The most significant difference is between the movie with NC-17 & R rated movies. There p-value is 0.71**

### Visual

**We can see that PG-13 movies got the highest amount of revenue out of all the certification types.**

### Conclusion
**We reject the null hypothesis**
**Therefore ther is a significant difference between the certification of a movie and the amount of revenue that it generates.**
