```python
import glob

q = 'Data/_1st_*.csv.gz'

file_list = sorted(glob.glob(q))

file_list
```

```python
#concatinating all the datasets
df = pd.concat([pd.read_csv(f, lineterminator = '\n') for f in file_list])

df
```
